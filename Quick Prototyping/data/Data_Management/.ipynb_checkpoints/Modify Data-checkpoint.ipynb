# Modify the Vaccinations Data

```python
filepath = '../Complete/country_vaccinations.csv'

df = pd.read_csv(filepath) 
df
```

```python
dictionary = {}
for index, row in df.iterrows():
    if(dictionary.__contains__(row.country)):
        if(math.isnan(dictionary[row.country])):
            dictionary[row.country] = row.total_vaccinations
        dictionary[row.country] += row.total_vaccinations
    else:
        dictionary[row.country] = row.total_vaccinations
        
iso_code = df['iso_code'].unique()
print(len(iso_code) == len(dictionary))

data = []
for i, key in enumerate(dictionary.keys()):
    data.append((key, iso_code[i], dictionary[key]))
    
df2 = pd.DataFrame(data, columns=["Name", "ISO_Code", "Total_Vaccinations"])
df2
```

```python
df2.to_csv('../Modified/Total_Vaccinations.csv', index=False)
```

# Modify the Covid Cases Data

```python
filepath = '../Complete/owid-covid-data.csv'

df = pd.read_csv(filepath) 
df
```

```python
dictionary = {}
for index, row in df.iterrows():
    if(dictionary.__contains__(row.location)):
        if(math.isnan(dictionary[row.location])):
            dictionary[row.location] = row.total_cases
        dictionary[row.location] += row.total_cases
    else:
        dictionary[row.location] = row.total_cases
        
iso_code = df['iso_code'].unique()
print(len(iso_code) == len(dictionary))

data = []
for i, key in enumerate(dictionary.keys()):
    data.append((key, iso_code[i], dictionary[key]))
    
df2 = pd.DataFrame(data, columns=["Name", "ISO_Code", "Total_Cases"])
df2
```

```python
df2.to_csv('../Modified/Total_Cases.csv', index=False)
```
